```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to change these)
city_file = "city_data.csv"
ride_file = "ride_data.csv"
```

```python
city = pd.read_csv(city_file)
ride = pd.read_csv(ride_file)
```

```python
data = pd.merge(ride,city, on = 'city')
data.head()
```

```python
a = data.groupby(['city']).mean()
b = data.groupby(['city']).count()

data_m = pd.merge(a,b, on = 'city',suffixes = [' mean',' count']).reset_index()
data_type = data.loc[:,["city","type"]]
data_type
data_m_t = pd.merge(data_m,data_type,on = "city").drop_duplicates(['city'],keep='first').set_index("city")
data_m_t.head()
```

```python
urban = data_m_t.loc[data_m_t["type_y"] == "Urban",["type_y","ride_id count","fare mean","driver_count mean"]]
suburban = data_m_t.loc[data_m_t["type_y"] == "Suburban",["type_y","ride_id count","fare mean","driver_count mean"]]
rural = data_m_t.loc[data_m_t["type_y"] == "Rural",["type_y","ride_id count","fare mean","driver_count mean"]]
```

```python
#Bubbles
plt.scatter(urban['ride_id count'],urban['fare mean'],s=urban['driver_count mean']*10,c='red',alpha=0.7,edgecolors='black',label='Urban')
plt.scatter(suburban['ride_id count'],suburban['fare mean'],s=suburban['driver_count mean']*10,c='blue',alpha=0.7,edgecolors='black',label='Suburban')
plt.scatter(rural['ride_id count'],rural['fare mean'],s=rural['driver_count mean']*10,c='yellow',alpha=0.7,edgecolors='black',label='Rural')

#name n grid

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Numbers of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()
plt.legend()


plt.show()
```

# Total Fares by City Type

```python
city_type = data.groupby(['type']).sum().reset_index()
f = city_type['fare']
explode = [0,0,0.1]
l = city_type['type']
colors = ['yellow','#a2cffe','#ff474c']
plt.title("% of Total Fares by City Type")
plt.pie(f, explode=explode, labels=l, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=145)
plt.axis("equal")
plt.show()
```

# Total Rides by City Type

```python
city_type_r = data.groupby(['type']).count().reset_index()
f = city_type_r['ride_id']
explode = [0,0,0.1]
l = city_type_r['type']
colors = ['yellow','#a2cffe','#ff474c']
plt.title("% of Total Rides by City Type")
plt.pie(f, explode=explode, labels=l, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=145)
plt.axis("equal")
plt.show()
```
