```python
import pandas as pd
import numpy as np
```

```python
file = 'purchase_data.csv'
hop = pd.read_csv(file) 
```


# Total Players

```python
TP=len(hop['SN'].unique())
pd.DataFrame({'Total Players':[TP]})
```

# Purchasing Analysis

```python
TPurchase=len(hop['Item ID'].unique())
APurchase=hop['Price'].mean()
CPurchase=hop['Item Name'].count()
Rev=hop['Price'].sum()
PA=pd.DataFrame({'Number of Unique Items':[TPurchase],'Average Purchase Price':[APurchase],
                'Total Number of Purchases':CPurchase,'Total Revenue':[Rev]})
format_m={'Average Purchase Price':'${:,.2f}',
          'Total Revenue':'${:,.2f}'}
for key, value in format_m.items():
    PA[key] = PA[key].apply(value.format)
PA
```

#  Gender Demographics

```python
hop_sn= hop.drop_duplicates(['SN'],keep='first')
count_sex = pd.DataFrame(hop_sn['Gender'].value_counts())
tt=count_sex['Gender'].sum()
count_sex['percentage'] = count_sex['Gender']/tt*100
count_sex['percentage'] = count_sex['percentage'].apply('{:.2f}%'.format)
count_sex
```

```python
#crear DataFrame con SN Ãºnicos
hopsn_gr = hop.groupby(['SN'])
c=(pd.DataFrame(hopsn_gr['Purchase ID','Item ID','Item Name'].count())).reset_index()
m=(pd.DataFrame(hopsn_gr['Age'].mean())).reset_index()
s=(pd.DataFrame(hopsn_gr['Price'].sum())).reset_index()
hop_unique_sn = pd.merge(pd.merge(c,m,on ='SN'),s,on='SN')
hop2=hop.loc[:,['SN','Gender']]
hop_unique_sn = pd.merge(hop_unique_sn,hop2,on='SN')
hop_unique_sn= hop_unique_sn.drop_duplicates(['SN'],keep='first')
hop_unique_sn.head()
```

# Purchasing Analysis (Gender)

```python
hop_g = hop_unique_sn.groupby(['Gender'])
pc=pd.DataFrame(hop_g['Purchase ID'].sum())
avgpppp=pd.DataFrame(hop_g['Price'].mean())
tpv=pd.DataFrame(hop_g['Price'].sum())
PurchasingAnalysis = pd.merge(pd.merge(pc,avgpppp,on='Gender'),tpv,on='Gender')
PurchasingAnalysis['Average Purchase Price'] = PurchasingAnalysis['Price_y']/PurchasingAnalysis['Purchase ID']
#Rename n reorder columns
PurchasingAnalysis = PurchasingAnalysis.rename(columns={'Purchase ID':'Purchase Count',
                                                       'Price_x':'Avg Total Purchase per Person',
                                                       'Price_y':'Total Purchase Value',
                                                       })
PurchasingAnalysis = PurchasingAnalysis[['Purchase Count','Average Purchase Price','Total Purchase Value',
                                        'Avg Total Purchase per Person']]
#format
format_m={'Average Purchase Price':'${:,.2f}',
          'Avg Total Purchase per Person':'${:,.2f}'}
for key, value in format_m.items():
    PurchasingAnalysis[key] = PurchasingAnalysis[key].apply(value.format)

PurchasingAnalysis
```

#  Age Demographics

```python
max_age = hop_unique_sn['Age'].max()
bins = [0, 9, 14, 19, 24, 29, 34, 39, max_age]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
hop_unique_sn["Age Range"] = pd.cut(hop_unique_sn['Age'],bins,labels=labels)
###
hop_age = hop_unique_sn.groupby(['Age Range'])
age_range = (pd.DataFrame(hop_age['SN'].count()))
age_range['Percentage of Players'] = age_range['SN']/tt*100
age_range['Percentage of Players'] = age_range['Percentage of Players'].apply('{:.2f}%'.format)
age_range

```

#  Purchasing Analysis (Age)

```python
pc=pd.DataFrame(hop_age['Purchase ID'].sum())
avgpppp=pd.DataFrame(hop_age['Price'].mean())
tpv=pd.DataFrame(hop_age['Price'].sum())
PurchasingAnalysisage = pd.merge(pd.merge(pc,avgpppp,on='Age Range'),tpv,on='Age Range')
PurchasingAnalysisage['Average Purchase Price'] = PurchasingAnalysisage['Price_y']/PurchasingAnalysisage['Purchase ID']
PurchasingAnalysisage = PurchasingAnalysisage.rename(columns={'Purchase ID':'Purchase Count',
                                                              'Price_x':'Avg Total Purchase per Person',
                                                              'Price_y':'Total Purchase Value'})
PurchasingAnalysisage = PurchasingAnalysisage[['Purchase Count','Average Purchase Price','Total Purchase Value',
                                               'Avg Total Purchase per Person']]
#format
format_m={'Average Purchase Price':'${:,.2f}',
          'Avg Total Purchase per Person':'${:,.2f}'}
for key, value in format_m.items():
    PurchasingAnalysisage[key] = PurchasingAnalysisage[key].apply(value.format)
    
PurchasingAnalysisage
```

#  Top Spenders

```python
hopsn=hop.groupby(['SN'])
Top_Spenders=pd.DataFrame(hopsn.sum().nlargest(5,['Price']))
Top_Spenders=Top_Spenders.loc[:,['Price']]
pc=pd.DataFrame(hopsn.count())
avg=pd.DataFrame(hopsn.mean())
Top_Spenders =pd.merge(pd.merge(pc,avg,on='SN'),Top_Spenders,on='SN')
Top_Spenders = Top_Spenders[['Purchase ID_x','Price_y','Price']]
Top_Spenders = Top_Spenders.rename(columns={'Purchase ID_x':'Purchase Count','Price_y':'Average Purchase Price',
                                            'Price':'Total Purchase Value'
                                           })
Top_Spenders = Top_Spenders.sort_values(by=['Total Purchase Value'],ascending=False)
#format
format_m={'Average Purchase Price':'${:,.2f}',
          'Total Purchase Value':'${:,.2f}'}
for key, value in format_m.items():
    Top_Spenders[key] = Top_Spenders[key].apply(value.format)

Top_Spenders 
```

#  Most Popular Items

```python
mpi=hop.groupby(['Item ID','Item Name'])
a=pd.DataFrame(mpi.count().nlargest(5,['Purchase ID']))
b=pd.DataFrame(mpi.mean())
c=pd.DataFrame(mpi.sum())
Popular_Items = pd.merge(a,pd.merge(b,c,on=['Item ID','Item Name']),on=['Item ID','Item Name'])
Popular_Items = Popular_Items.rename(columns={'Purchase ID':'Purchase Count','Price_x':'Item Price',
                                              'Price_y':'Total Purchase Value'})
#format
format_m={'Item Price':'${:,.2f}',
          'Total Purchase Value':'${:,.2f}'}
for key, value in format_m.items():
    Popular_Items[key] = Popular_Items[key].apply(value.format)

Popular_Items = Popular_Items.loc[:,['Purchase Count','Item Price','Total Purchase Value']]
Popular_Items

```

#  Most Profitable Items

```python
c=pd.DataFrame(mpi.count())
b=pd.DataFrame(mpi.mean())
a=pd.DataFrame(mpi.sum().nlargest(5,['Price']))
P_Items = pd.merge(a,pd.merge(b,c,on=['Item ID','Item Name']),on=['Item ID','Item Name'])
P_Items = P_Items.rename(columns={'Purchase ID_y':'Purchase Count','Price_x':'Item Price',
                                              'Price':'Total Purchase Value'})
#format
format_m={'Item Price':'${:,.2f}',
          'Total Purchase Value':'${:,.2f}'}
for key, value in format_m.items():
    P_Items[key] = P_Items[key].apply(value.format)

P_Items=P_Items.loc[:,['Purchase Count','Item Price','Total Purchase Value']]
P_Items

```
